The dataset is in a csv file in the homework's repo (salmon_data.csv) and has the following columns:

```python
### ===== FINAL CODE =====

#import the relevant libraries
import pandas as pd

#call the data that is already in the repo
catch_original = pd.read_csv('salmon_data.csv')

#update the dataframe using drop so it doesn't have the notesRegCode column
catch_original.drop(columns = ['notesRegCode'], inplace = True)

#lets update the value to 1 using at method
catch_original.at[400, 'Catch'] = 1

#lets try updating the col again using same astype method
catch_original['Catch'] = catch_original.Catch.astype('int64')

#use the mean function to find the mean catch per region
average_catch = catch_original.groupby('Region').mean()

# solve exercise here
average_catch.sort_values('Catch', inplace = True)

average_catch.Catch.plot(kind = 'bar',
                        xlabel = 'Region',
                        ylabel = 'Average catch',
                        title = 'Estimated Average catch per region between 1878 -1997',
                        color = 'pink')
```
